### Example 10.3:
We have for given scenario:

n = 100; $\overline{x}$ = 71.8; $\sigma$ = 8.9; $\alpha$ = 0.05

Now, for the hypothesis we have:

H$_o$: $\mu$ = 70 years ; H$_1$: $\mu$ > 70 years

Now, for the test static, we have:

$z$ = $(\overline{x} - \mu)/(\sigma/\sqrt{n})$ = (71.8 - 70)/(8.9/$\sqrt{100}$) = 1.8/0.89 = 2.02

__Using Critical Values:__
Now, for critical region, we have to apply for the right side i.e.

$z$ > $z_{\alpha}$ => $z$ > $z_{0.05}$ => $z$ > ((1.64 + 1.65)/2) => $z$ > 1.645

Since, we have 2.02 > 1.645. Thus, the value lies in critical region i.e. we reject the null hypothesis and accept the alternative. Hence, we conclude that the mean life span today is greater tahn 70 years. 

__Using P Value:__
For P-value, the value for the computed test statistic can be given using the provided table as:

p = $P(Z > 2.02)$ = 1 - $P(Z < 2.02)$ = 1 - 0.9783 = 0.0217

Now since we have: p < $\alpha$ , thus, the test static lies in the critical region i.e. we reject the null hypothesis.

### Example 10.4:

$z$ < $z_{-\alpha/2}$ => $z$ < $z_{-0.01/2}$ => $z$ < $z_{-0.005}$ => $z$ < ((-2.57 - 2.58)/2) => $z$ < -2.575

Since we have -2.83 < -2.575, thus, the test static lies in the critical region i.e. we reject the null hypothesis and accept the alternatie hypothesis. We conclude that average breaking strength is not equal to 8 but is in fact, less than 8 kilograms

__Using P Value:__
p = $P(|Z| > 2.83)$ = 2$P(Z < -2.83)$ = 2(0.0023) = 0.0046

Since we have: p < $\alpha$, thus, the test static lies in the critical region i.e. we reject the null hypothesis.

### Example 10.6:
For given case scenario, we have:

$\overline{x_1}$ = 85; $\sigma_{1}$ = 4; n$_1$ = 12 ; $\overline{x_2}$ = 81; $\sigma_{2}$ = 5; n$_2$ = 10; $\alpha$ = 0.05

For hypothesis, we have:

H$_o$ : $\mu_{1}$ - $\mu_{2}$ = 2 units ; H$_1$ : $\mu_{1}$ - $\mu_{2}$ > 2

Now, for test static, we have to apply pooled t-test. First finding the pooled standard deviation:

$s_p$ = $\sqrt{(s_{1}^{2}(n_1 - 1) + s_{2}^{2}(n_2 - 1))/(n_1 + n_2 - 2)}$ = $\sqrt{((16)(12-1) + (25)(10-1))/(12+10-2)}$ = $\sqrt{401/20}$ = $\sqrt{20.05}$ = 4.478

Now, for test static:

$t$ = $((\overline{x_1} - \overline{x_2}) - d_o)/(s_p)(\sqrt{(1/(n_1) + 1/(n_2))})$ = (85-81-2)/(4.478)($\sqrt{(1/12 + 1/10)}$) = (2)/(4.478)(0.428) = 2/1.916 = 1.04

__Using P Values:__
For the given pooled t-test, we have: $\upsilon$ = n$_1$ + n$_2$ - 1 = 12 + 10 - 1 = 21

p = $P(T > 1.04 | \upsilon = 21)$  = 0.155 $\approx$ 0.16 

Now, since we have p > $\alpha$ , thus, we reject the null hypothesis and conclude that the abrasive wear of material exceeds that of material 2 by more than 2 units
