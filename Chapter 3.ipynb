### Example 3.9:

### Example 3.10:

$\int_{-\infty}^{\infty} f(x) dx$ = 1

Now, we have:

### Example 3.12:
To find the CDF for the function in the above example, we set lower limit to a value x and upper limit to $\infty$, i.e. we have:

$\int_{-\infty}^{x} f(x) dx$ = $\int_{-1}^{x}$ (x$^2$/3) $dx$ = (x$^3$/9) |$_{-1}^{x}$ = x$^3$/9 + 1/9 = (x$^3$ + 1)/9 

Now for P(0 $<$ X $\leq$ 1) , we have:

P(0 $<$ X $\leq$ 1) = F(1) - F(0) = (1$^3$ + 1)/9 - (0$^3$ + 1)/9 = 2/9 - 1/9 = 1/9
