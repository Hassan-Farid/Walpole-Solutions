### Example 2.17:
In this problem, we can have two cases for four digit even numbers i.e. 

(a) The last digit is zero 
(b) The last digit is 2 or 6

(a) In case the last digit is zero, then we have 1 possibility for unit number, 5 possibilities for tens, 4 possibilities for hundred and 3 possibilities for thousand i.e. O(S) = $n_1 * n_2 * n_3 * n_4$ = 1 * 5 * 4 * 3 = 60 possibilites

(b) In case the last digit is not zero, then we have 2 possibilites for unit number (either 2 or 6), 4 possibilites for thousand (since 0 cannot be used in thousand place otherwise number will become 3 digit), 4 possibilities for hundred place and 3 possibilites for tens place, i.e. O(S) = $n_1 * n_2 * n_3 * n_4$ = 2 * 4 * 4 * 3 = 96 possibilites

Combining these two, we will get the final result i.e. O(S) = 60 + 96 = 156 possibilities

__Note:__ In these type of questions, where conditions are specified, we prioritize conditions over the default order of possibilities e.g. one might have for (b) part consider putting 2 possibilites in unit place (due to 2 or 6) and then putting 5 in place of tens place.. but we didn't do that because we had another condition that 0 could not be used in the thousands place. SO we first considered the thousand place possibilities and then moved to hundred. Notice that since the conditions were only two, thus, switching the possibilites between the hundred and tens place would not change the result in anyway i.e. if we put $n_3$ = 3 and $n_4$ = 4, it does not change the final result. 

### Example 2.18:
The multiplication rule of counting can be summarized to a permutation formula if and only if the order for the items does not matter and repetition is not allowed, otherwise, always use the multiplication formula. Here, it is being told that 3 awards are to be given to 25 graduate student with each student getting only one award at a time (i.e. repetition is not allowed and it does not matter whether the 5th student gets the award or the 22nd), hence, we can apply formula for permutation for the given event as: O(A) = $^nP_r$ = $^25P_3$ = 25!/(25-3)! = 25!/22! = (25)(24)(23) = 13,800 possible selections (Same as applying multiplication rule of counting but simpler to evaluate)

### Example 2.19:
__(a)__ For the first case, there are no restrictions i.e. anyone can become a president and treasurer, hence, applying formula for permutation: O(S) = $^nP_r$ = $^50P_2$ = 2450 choices

__(b)__ For the second case, it is the condition that A will only serve as officer if he is the president otherwise not. Hence we can have two sub-cases here i.e. 

(i) A is the president: then the remaining 49 can become the treasurer, thus we have: O($A_1$) = 49
(ii) A is not the president: then he will not serve as an officer and the total number of officers reduce from 50 to 49. Now from these 49 there are no restrictions, hence applying formula for permutation: O($A_2$) = $^nPr$ = $^49P_2$ = 2352 

Thus, total choices for this case become O(A) = 2352 + 49 = 2401 choices

__(c)__ For the third case, it is the condition that B and C will either serve together or not at all. Hence, we can have two sub-case here i.e.

(i) B and C both serve: then they both occupy the position of the leader and treasurer i.e. O($B_1$) = 2
(ii) B and C both do not serve: then the total number of officers to select from will be 48. Hence, we can apply the permutation formula on the remaining officers as: O($B_2$) = $^nP_r$ = $^48P_2$ = 2256

Thus, the total choices for the third case become O(B) = 2 + 2256 = 2258 choices

__(d)__ For the fourth case, it is given that D and E will not serve together, this can be computed by first computing the possibility that when D and E will serve together and take the compliment of that event. Hence, we have for D and E serving together: O(C) = 2 (as both will take either treasurer or leader's place). Now we have, O(C') = O(S) - O(C) = 2450 -2 = 2448 choices
